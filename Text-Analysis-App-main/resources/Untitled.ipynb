```python
import pandas as pd

first = pd.read_excel("華語八千詞表20200917_整理.xlsx", sheet_name="準備級一級")
second = pd.read_excel("華語八千詞表20200917_整理.xlsx", sheet_name="準備級二級")
elementary = pd.read_excel("華語八千詞表20200917_整理.xlsx", sheet_name="入門級")
basic = pd.read_excel("華語八千詞表20200917_整理.xlsx", sheet_name="基礎級")
advanced = pd.read_excel("華語八千詞表20200917_整理.xlsx", sheet_name="進階級")
higher = pd.read_excel("華語八千詞表20200917_整理.xlsx", sheet_name="高階級")
fluence = pd.read_excel("華語八千詞表20200917_整理.xlsx", sheet_name="流利級")
```

```python
first_list = list(first["詞彙"])
first_list += [" " for i in range(len(fluence)-len(list(first["詞彙"])))]

second_list = list(second["詞彙"])
second_list += [" " for i in range(len(fluence)-len(list(second["詞彙"])))]

elementary_list = list(elementary["詞彙"])
elementary_list += [" " for i in range(len(fluence)-len(list(elementary["詞彙"])))]

basic_list = list(basic["詞彙"])
basic_list += [" " for i in range(len(fluence)-len(list(basic["詞彙"])))]

advanced_list = list(advanced["詞彙"])
advanced_list += [" " for i in range(len(fluence)-len(list(advanced["詞彙"])))]

higher_list = list(higher["詞彙"])
higher_list += [" " for i in range(len(fluence)-len(list(higher["詞彙"])))]
```

```python
df = pd.DataFrame({"準備級一級": first_list, "準備級二級": second_list,
                   "入門級": elementary_list, "基礎級": basic_list, 
                   "進階級": advanced_list, "高階級": higher_list, 
                   "流利級": list(fluence["詞彙"])})

df.to_csv("華語八千詞表20220314_整理.csv", encoding="utf-8", index=False)
```
